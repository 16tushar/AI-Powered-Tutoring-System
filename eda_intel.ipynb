```python
import random
import faker
import pandas as pd

# Initialize Faker for generating fake data
fake = faker.Faker()

def generate_k12_profile():
    """Generates a single K-12 student profile with randomized, unique data."""
    gender = random.choice(['Male', 'Female', 'Other'])
    age = random.randint(5, 18)  # K-12 age range
    grade = age - 5
    region = random.choice(['North', 'South', 'East', 'West', 'Central'])

    # Adjust score and IQ to match grade level to avoid bias
    # Students in lower grades typically have lower scores and IQ, but this can vary.
    if grade <= 5:
        score = random.randint(50, 85)
        iq = random.randint(80, 110)
    elif grade <= 8:
        score = random.randint(60, 90)
        iq = random.randint(85, 115)
    else:
        score = random.randint(70, 95)
        iq = random.randint(90, 120)

    return {
        'Name': fake.name(),
        'Gender': gender,
        'Phone Number': fake.phone_number(),
        'Age': age,
        'Region': region,
        'Grade': grade,
        'Score': score,
        'IQ': iq
    }

def generate_k12_dataset(num_profiles):
    """Generates a list of unique K-12 student profiles."""
    profiles = []
    phone_numbers = set()  # To ensure unique phone numbers

    while len(profiles) < num_profiles:
        profile = generate_k12_profile()
        if profile['Phone Number'] not in phone_numbers: #Enforce unique phone numbers.
            profiles.append(profile)
            phone_numbers.add(profile['Phone Number'])

    return profiles

def main():
    """Generates the dataset and saves it to a CSV file."""
    num_profiles = 500
    dataset = generate_k12_dataset(num_profiles)

    # Convert to Pandas DataFrame and save to CSV
    df = pd.DataFrame(dataset)
    df.to_csv('k12_student_profiles.csv', index=False)  # Save to CSV

    print(f"Successfully generated {num_profiles} unique K-12 student profiles and saved to k12_student_profiles.csv")

if __name__ == "__main__":
    main()

```
